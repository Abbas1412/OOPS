```python
''' Problem 1: Bank Account Create a class representing a bank account with attributes like account number,
account holder name, and balance. Implement methods to deposit and withdraw money from the account.'''

#Code:-

class Bank_Account:
    
    def __init__(self, account_number, account_holder_name, balance=0.0):
        self.account_number = account_number
        self.account_holder_name = account_holder_name
        self.balance = balance
    
    def deposit(self):
        try:
            amount = float(input("Enter amount you want to deposit: "))
            
            if amount > 0:
                self.balance += amount
                print("Deposited Successfully.")
            else:
                print("Amount should be greater than zero.")
        except ValueError as e:
            print(e)
    
    def withdraw(self):
        try:
            amount = float(input("Enter amount you want to withdraw: "))
            
            if self.balance >= amount and amount > 0:
                self.balance -= amount
                print(f"You have withdrawn {amount} Rs successfully.\n")
            elif amount <= 0:
                print("Withdrawal amount should be greater than zero.\n")
            else:
                print("You don't have enough balance.\n")
        except ValueError as e:
            print(e)

    def get_balance(self):
        return self.balance

    def __str__(self):
        return f"Account Number: {self.account_number}\nAccount Holder: {self.account_holder_name}\nBalance: {self.balance:.2f}"


account1 = Bank_Account("12345", "Zamin Abbas", 1000.0)

account1.withdraw()

account1.deposit()

account1.get_balance()
```

```python
'''Problem 2: Employee Management Create a class representing an employee with attributes like employee ID, name, and salary.
Implement methods to calculate the yearly bonus and display employee details.'''

#Code:-

import logging

class Employe:
    
    def __init__(self, employee_id, employee_name, employee_salary):
        
        self.employee_id = employee_id
        self.employee_name = employee_name
        self.employee_salary = employee_salary
        
        
    def calculate_yearly_bonus(self):
        
        try:
            #Assuming yearly bonus of 10 %
            yearly_bonus = 0.10 * self.employee_salary
            return yearly_bonus
            
        except Exception as e:
            logging.error(f"Error in calculating the bonus of employee {self.employee.id} : {str(e)}")
            
                          
                
    def display_employee_details(self):
        
        try:
            yearly_bonus = self.calculate_yearly_bonus()
            print(f"Employe ID: {self.employee_id}\nEmploye Name: {self.employee_name}\nEmploye Salary: ${self.employee_salary:.2f}\nYearly Bonus: ${yearly_bonus:.2f}")
        
        except Exception as e:
            logging.error(f"Error in printing the values: {str(e)}")
            
    

emp1 = Employe(12,"Abhi",1233)

emp1.display_employee_details()
```

```python
'''Problem 3: Vehicle Rental Create a class representing a vehicle rental system.
Implement methods to rent a vehicle, return a vehicle, and display available vehicles.'''

#Code:-

import logging

class Vehicle_Rental_System:
    
    def __init__(self):
        self.available_vehicles = {}
        self.rental_vehicles = {}
    
    
    def add_vehicle(self, vehicle_id, vehicle_type):
        try:
            if(vehicle_id not in self.available_vehicles):
                self.available_vehicles[vehicle_id] = vehicle_type
                
        except Exception as e:
            logging.info(e)
            
            
    def remove_vehicle(self, vehicle_id):
        try:
            
            if vehicle_id not in self.available_vehicles:
                raise ValueError(f"Vehicle with ID: {vehicle_id} not present in the system")

            del self.available_vehicles[vehicle_id]
            logging.info(f"Vehicle with ID: {vehicle_id} removed from the system")
            
        except Exception as e:
            logging.info(e)
            
            
            
    def rent_vehicle(self, vehicle_id, customer_name):
        
        try:
            if vehicle_id not in self.available_vehicles:
                raise ValueError(f"Vehicle with ID {vehicle_id} not available:")


            if vehicle_id in self.rental_vehicles:
                raise ValueError(f"Vehicle with ID {vehicle_id} already rented:")

            else:
                self.rental_vehicles[vehicle_id] = self.available_vehicles[vehicle_id]
                self.remove_vehicle(vehicle_id)
                logging.info(f"Vehicle of id: {vehicle_id} is rented to {customer_name}")
                
        except Exception as e:
            logging.info(f"{e}")
            
        
    
    def return_vehicle(self, vehicle_id, customer_name):
        try:
        
            if vehicle_id in self.available_vehicles:
                raise ValueError(f"Vehicle with ID {vehicle_id} already present in the system")

            if vehicle_id not in self.rental_vehicles:
                raise ValueError(f"Vehicle with ID {vehicle_id} not rented")

            else:
                self.add_vehicle(vehicle_id, self.rental_vehicles[vehicle_id])
                del self.rental_vehicles[vehicle_id]
                
        except Exception as e:
            logging.info(e)

        
    def display_available_vehicles(self):
        try:
            logging.info(f"Available vehicles in the system are {self.available_vehicles.items()}")
            
        except Exception as e:
            logging.error(e)
            
    def display_rental_vehicles(self):
        try:
            logging.info(f"Rental vehicles from the system are {self.rental_vehicles.items()}")
            
        except Exception as e:
            logging.error(e)

logging.basicConfig(level = logging.INFO)

rental_system = Vehicle_Rental_System()

rental_system.add_vehicle("001", "Sedan")
rental_system.add_vehicle("002", "SUV")
rental_system.add_vehicle("003", "Truck")

rental_system.display_available_vehicles()

rental_system.rent_vehicle("001", "John")
rental_system.return_vehicle("001", "John")

rental_system.display_available_vehicles()

rental_system.display_rental_vehicles()
```

```python
''' Problem 4: Library Catalog Create classes representing a library and a book.
   Implement methods to add books to the library, borrow books, and display available books.'''

#Code:-

import logging
logging.basicConfig(filename = 'library.log', level = logging.INFO, format = '%(asctime)s - %(levelname)s - %(message)s')

class Book:
    
    def __init__(self, title, author):
        
        self.title = title
        self.author = author
        self.available = True
        
    def __str__(self):
        return f"{self.title} by {self.author}"


class Library:
    
    def __init__(self):
        self.books = []
    
    def add_book(self, book):
        
        if(isinstance(book,Book)):
            
            if book not in self.books:
                self.books.append(book)
                logging.info(f"\nBook with title {book.title} by {book.author} added to library")
                
            else:
                logging.error(f"\nBook with title {book.title} by {book.author} already added to library")
        else:
            raise ValueError(f"\nInvalid Book instance")
    
    
    def remove_book(self, book):
        
        if isinstance(book,Book):
            
            if book in self.books:
                self.books.remove(book)
                logging.info(f"Book with title {book.title} by {book.author} removed from library")
                
            else:
                logging.error(f"Book with title {book.title} by {book.author} not present in library")
            
        else:
            raise ValueError(f"Invalid book instance")
        
    
    def borrow_book(self, book_title):
            
        for book in self.books:
                
            if book_title.lower() == book.title.lower() and book.available is True:
                book.available = False
                logging.info(f"book with title{book.title} by author {book.author} is borrowed from library")
                return book
            
        raise ValueError(f"book with title {book_title} not available")
        
    
    def display_available_books(self):
        
        count = 0
        for book in self.books:
            
            if book.available == True:
                logging.info(f"{book.title} by {book.author}\n")
                count += 1
        if(count == 0):
            raise ValueError(f"No book available in library")

        

if __name__ == "__main__":
    
    library = Library()

    while True:
        print("\nLibrary Menu\n")
        print("1. Add Book\n2. Borrow Book\n3. Remove Book\n4. Show Available Books\n5. Exit\n")

        try:
            choice = int(input("Enter Your Choice (1-5): "))

            if choice == 1:
                book = Book(input("\nEnter Book Title: "), input("Enter Author Name: "))
                library.add_book(book)

            elif choice == 2:
                library.borrow_book(input("\nEnter Book Title: "))

            elif choice == 3:
                book = Book(input("\nEnter Book Title: "), input("Enter Author Name: "))
                library.remove_book(book)

            elif choice == 4:
                library.display_available_books()

            elif choice == 5:
                break

        except Exception as e:
            logging.error(f"An error occurred: {e}")
```

```python
''' Problem 5: Product Inventory Create classes representing a product and an inventory system.
Implement methods to add products to the inventory, update product quantity, and display available products.'''

#Code:-

import logging
logging.basicConfig(filename = 'library.logging', level = logging.INFO, format = '%(asctime)s - %(levelname)s - %(message)s')

class Product:
    
    def __init__(self,product_id, name, price, stock_quantity):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.stock_quantity = stock_quantity
        
    def __str__(self):
        return f"Product_id: {self.product_id}\nName: {self.name}\nPrice: {self.price}\nstock_quantity: {self.stock_quantity}\n\n"


class Inventory_System:
    
    def __init__(self):
        self.inventory = {}
    
    
    def add_product(self, product):
        
        if isinstance(product,Product):
            
            if product.product_id in self.inventory:
                self.inventory[product.product_id].stock_quantity += product.stock_quantity
                logging.info(f"\n{product} added to inventory")
                             
            else:
                self.inventory[product.product_id] = product
                logging.info(f"\n{product} added to inventory")
                
        else:
            raise ValueError(f"\nInvalid Product Object")
                
        
    def update_quantity(self, product_id, quantity):
        
        if quantity >= 0:
            
            if product_id in self.inventory:
                prev_quantity = self.inventory[product_id].stock_quantity
                self.inventory[product_id].stock_quantity = quantity
                logging.info(f"\nProduct Quantity updated from {prev_quantity} to {quantity}")
                    
            else:
                 raise ValueError(f"Product not present in the inventory")
                    
                 
        else:
            raise ValueError("\nQuantity should be greater than zero")
            
            
            
    def display_available_products(self):
        
        available_products = [product for product in self.inventory.values() if product.stock_quantity > 0]
        
        if available_products:
            for product in available_products:
                logging.info(str(product))

        else:
            print(f"No product is available in inventory")
            
            
if __name__ == "__main__":
    
    inventory_system = Inventory_System()

    # Adding products to the inventory
    product1 = Product(1, "Laptop", 999.99, 10)
    product2 = Product(2, "Smartphone",  699.99, 15)
    inventory_system.add_product(product1)
    inventory_system.add_product(product2)

    # Updating product quantity
    try:
        inventory_system.update_quantity(1, 5)
    except ValueError as e:
        print(e)

    # Displaying available products
    inventory_system.display_available_products()
```

```python
'''Problem 7: Student Management Create a class representing a student with attributes like student ID, name, and grades.
Implement methods to calculate the average grade and display student details.'''

#Code:-

import logging
logging.basicConfig(filename = 'test.log', level = logging.INFO, format = '%(asctime)s - %(levelname)s - %(message)s')


class Student:
    
    def __init__(self, student_id, name):
        
        self.student_id = student_id
        self.name = name
        self.grades = []
                    
                    
    def add_grade(self, grade):
                    
        if 0 <= grade <= 100:
            self.grades.append(grade)
            logging.info(f"Grade added to the list")
            
        else:
            raise ValueError(f"\nInvalid Grade")

        
    
    def calculate_average_grade(self):
        
        if not self.grades:
            raise ValueError("No Grades are available")
                    
        else:
            return sum(self.grades)/len(self.grades)
            
        
        
    def display_student_details(self):
        
        try:
            average_grade = self.calculate_average_grade()
            logging.info(f"Student ID: {self.student_id}\nName: {self.name}\nAverage Grade: {average_grade}")
            
        except Exception as e:
            logging.info(e)
            

            
if __name__ == '__main__':
    
     # Initialize a student
    student1 = Student(1, "John Doe")

    # Add grades to the student
    student1.add_grade(85)
    student1.add_grade(92)
    student1.add_grade(78)

    # Display student details
    logging.basicConfig(level=logging.INFO)
    student1.display_student_details()
```

```python
'''Problem 9: Social Media Profile Create a class representing a social media profile with attributes like username and posts.
Implement methods to add posts, display posts, and search for posts by keyword.'''

#Code:-

import logging
logging.basicConfig(filename ='test.log', level = logging.INFO, format = '%(asctime)s - %(levelname)s - %(message)s')


class SocialMediaProfile:
    
    def __init__(self, username):
        
        self.user_name = username
        self.posts = []
        
    
    def add_post(self, post_content):
        
        try:
            
            if post_content:
                posts.append(post_content)
                logging.info(f"post_content: {post_content} added to post")
                
                
            else:
                logging.warning("Empty post content post not added")
                
                
        except Exception as e:
            logging.error(e)
    
    def search_post(self, keyword):
        
        try:
            available_post = [post for post in self.posts if keyword in self.posts]
            
            if available_post:
                logging.info(f"Search Result {available_post}")
                
            else:
                raise ValueError("Post not Found")
                
        except Exception as e:
            logging.error(e)
        
    
    def display_posts(self):
        
        try:
            for post in self.posts:
                logging.info(post)
                
        except Exception as e:
            logging.error(e)
            
            
if __name__ == "__main__":
    username = "john_doe"
    profile = SocialMediaProfile(username)

    # Add some posts
    profile.add_post("Hello, world! This is my first post.")
    profile.add_post("Enjoying a beautiful day at the park.")
    profile.add_post("Exploring new places with friends.")

    # Display all posts
    profile.display_posts()

    # Search for posts by a keyword
    profile.search_post("park")
```
